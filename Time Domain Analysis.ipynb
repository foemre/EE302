
# Time Domain Analysis


Let's start with a simple example! We will investigate the effects of pole positions on the step response of a first order system.

The code below generates and plots the step response of a first order system. 
You can run the code and see the results yourself by using shift+enter or the play button if you download Jupyter notebook.


```python
# Here we import some necessary libraries 
from numpy import min
from scipy import linspace
from scipy.signal import lti, step
from matplotlib import pyplot as p

# Create an LTI transfer function from coefficients sys = tf([alpha],[1 alpha])
alpha=3;
tf = lti([alpha], [1, alpha])
# Compute the step response of the system
t, s = step(tf)

# Plotting stuff
p.plot(t, s)
p.grid()
p.xlabel('Time / s')
p.ylabel('Output')
p.show()
```

\begin{align}
G(s)=\frac{3}{s+3} \nonumber
\end{align}

Let's add some more plots and see the effects of changing $\alpha$. Let $\alpha$ be 3, 5 and 10 respectively.

```python
# Here we import some necessary libraries 
from numpy import min
from scipy import linspace
from scipy.signal import lti, step
from matplotlib import pyplot as p

# Create an LTI transfer function from coefficients sys = tf([alpha],[1 alpha])
alpha=3;
tf1 = lti([alpha], [1, alpha])
t1, s1 = step(tf1)

alpha=5;
tf2 = lti([alpha], [1, alpha])
t2, s2 = step(tf2)

alpha=10;
tf3 = lti([alpha], [1, alpha])
# Below call to the step(.) function also sets the time interval of the response computed
t3, s3 = step(tf3, T = linspace(0, 2, 200))  

# Plotting stuff
p.plot(t1, s1)
p.plot(t2, s2)
p.plot(t3, s3)
p.grid()
p.xlabel('Time / s')
p.ylabel('Output')
p.legend(['$\\alpha=3$', '$\\alpha=5$', '$\\alpha=10$'])
p.show()
```

Step response for the system:
\begin{align}
G(s)=\frac{\alpha}{s+\alpha}\quad\quad  for \quad \quad\alpha \in \{3, 5, 10 \}   \nonumber
\end{align}


Let's make $\alpha$ negative and see what happens to the step response!

```python
# Here we import some necessary libraries 
from numpy import min
from scipy import linspace
from scipy.signal import lti, step
from matplotlib import pyplot as p

# Create an LTI transfer function from coefficients sys = tf([alpha],[1 alpha])
alpha=-3;
tf = lti([1], [1, alpha])
t, s = step(tf)

# Plotting stuff
p.plot(t, s)
p.grid()
p.xlabel('Time / s')
p.ylabel('Output')
p.show()
```

Step response for the system:
\begin{align}
G(s)=\frac{1}{s-3}  \nonumber
\end{align}

#### The system becomes unstable!


## Second Order Systems


Let's start with a simple example! We will investigate the effects of pole positions on the step response of a secon order system.

The code below generates and plots the step response of a second order system. 
You can run the code and see the results yourself by using shift+enter or the play button 
\begin{align}
G(s)=\frac{\omega_n^2}{s^2+2\xi\omega_n +\omega_n^2} \nonumber
\end{align}




Let's start with observing the step response of undamped, underdamped, critically-damped and overdamped systems below.

```python
# Here we import some necessary libraries 
from numpy import min
from scipy import linspace
from scipy.signal import lti, step
from matplotlib import pyplot as p


# Create an LTI transfer function 
wn=2
ksi=0.0  # undamped 
tf1 = lti([wn*wn ], [1, 2*ksi*wn, wn*wn])
t1, s1 = step(tf1, T = linspace(0, 12.5, 200))  

ksi=0.2  # underdamped 
tf2 = lti([wn*wn ], [1, 2*ksi*wn, wn*wn])
t2, s2 = step(tf2, T = linspace(0, 12.5, 200))  

ksi=1  # critically damped 
tf3 = lti([wn*wn ], [1, 2*ksi*wn, wn*wn])
# Below call to the step(.) function also sets the time interval of the response computed
t3, s3 = step(tf3, T = linspace(0, 12.5, 200))  

ksi=3.  # overdamped 
tf4 = lti([wn*wn ], [1, 2*ksi*wn, wn*wn])
# Below call to the step(.) function also sets the time interval of the response computed
t4, s4 = step(tf4, T = linspace(0, 12.5, 200))  


# Plotting stuff
p.plot(t1, s1)
p.plot(t2, s2)
p.plot(t3, s3)
p.plot(t4, s4)
p.grid()
p.xlabel('Time / s')
p.ylabel('Output')
p.legend(['undamped', 'underdamped', 'critically damped','overdamped'])
p.show()
```

## Underdamped response
Let's see some examples of underdamped response below.

```python
# Here we import some necessary libraries 
from numpy import min
from scipy import linspace
from scipy.signal import lti, step
from matplotlib import pyplot as p


# Create an LTI transfer function 
wn=2
ksi=0.2 
tf1 = lti([wn*wn ], [1, 2*ksi*wn, wn*wn])
t1, s1 = step(tf1)

ksi=0.4
tf2 = lti([wn*wn ], [1, 2*ksi*wn, wn*wn])
t2, s2 = step(tf2)

ksi=0.6
tf3 = lti([wn*wn ], [1, 2*ksi*wn, wn*wn])
# Below call to the step(.) function also sets the time interval of the response computed
t3, s3 = step(tf3, T = linspace(0, 17.5, 200))  

ksi=0.8
tf4 = lti([wn*wn ], [1, 2*ksi*wn, wn*wn])
# Below call to the step(.) function also sets the time interval of the response computed
t4, s4 = step(tf4, T = linspace(0, 17.5, 200))  


# Plotting stuff
p.plot(t1, s1)
p.plot(t2, s2)
p.plot(t3, s3)
p.plot(t4, s4)
p.grid()
p.xlabel('Time / s')
p.ylabel('Output')
p.legend(['$\\xi=0.2$', '$\\xi=0.4$', '$\\xi=0.6$','$\\xi=0.8$'])
p.show()
```

Second order underdamped responses for different damping ratio values for the system:
\begin{align}
G(s)=\frac{\omega_n^2}{s^2+2\xi\omega_n +\omega_n^2},\quad \quad \text{where} \quad \quad \omega_n=2 \nonumber 
\end{align}
#### Notice that they have the same frequency!

Let's see some more examples:

```python
# Here we import some necessary libraries 
from numpy import min
from scipy import linspace
from scipy.signal import lti, step
from matplotlib import pyplot as p


# Create an LTI transfer function 
wn=2
ksi=0.3
tf1 = lti([wn*wn ], [1, 2*ksi*wn, wn*wn])
t1, s1 = step(tf1)

wn=4
tf2 = lti([wn*wn ], [1, 2*ksi*wn, wn*wn])
t2, s2 = step(tf2)

wn=8
tf3 = lti([wn*wn ], [1, 2*ksi*wn, wn*wn])
t3, s3 = step(tf3)


# Plotting stuff
p.plot(t1, s1)
p.plot(t2, s2)
p.plot(t3, s3)

p.grid()
p.xlabel('Time / s')
p.ylabel('Output')
p.legend(['$\\omega_n=2$', '$\\omega_n=4$', '$\\omega_n=8$'])
p.show()
```

Second order underdamped responses for different natural frequencies of the system:
\begin{align}
G(s)=\frac{\omega_n^2}{s^2+2\xi\omega_n +\omega_n^2},\quad \quad \text{where} \quad \quad \xi=0.3 \nonumber
\end{align}
#### Notice that they have the same overshoot!
