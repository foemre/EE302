## Examples

The code below generates and plots the step response of some second order systems. 
You can run the code and see the results yourself by using shift+enter or the play button if you download jupyter notebook.
\begin{align}
G(s)=\frac{\omega_n^2}{s^2+2\xi\omega_ns +\omega_n^2} \nonumber
\end{align}


```python
# Here we import some necessary libraries 
from scipy import linspace
from scipy.signal import lti, step
from matplotlib import pyplot as plt

```

### Example 1
Consider the following second order systems with different set of parameters:
\begin{align}
G_1(s)=\frac{\omega_1^2}{s^2+2 \xi_1 \omega_1 s +\omega_1^2}, \quad \quad G_2(s)=\frac{\omega_2^2}{s^2+2 \xi_2 \omega_2 s +\omega_2^2}.
\end{align}
The parameters of the systems are chosen as:
\begin{align}
\xi_1=0.2, \quad & w_1 =1.77, \nonumber \\
\quad \xi_2=0.5, \quad & w_2 =2.\nonumber
\end{align}
Compare the step responses of the two systems.

```python
ksi1=0.2; wn1=1.77
ksi2=0.5; wn2=2

# Create an LTI transfer function and plot step response
tf1 = lti([wn1*wn1 ], [1, 2*ksi1*wn1, wn1*wn1])
t1, s1 = step(tf1, T = linspace(0, 17.5, 200))  

tf2 = lti([wn2*wn2 ], [1, 2*ksi2*wn2, wn2*wn2])
t2, s2 = step(tf2, T = linspace(0, 17.5, 200))  

# Plotting pole locations and step responses
f, axarr = plt.subplots(1, 2)
f.set_size_inches(14, 4)
axarr[1].plot(t1, s1)
axarr[1].plot(t2, s2,'r')
axarr[1].grid()
axarr[1].legend(['system-1', 'system-2'])
axarr[1].set_xlabel('Time / s')
axarr[1].set_ylabel('Output')
axarr[1].set_title('Step Response')
axarr[0].plot([-ksi1*wn1, -ksi1*wn1 ], [wn1*(1-ksi1**2)**0.5, -wn1*(1-ksi1**2)**0.5], 'bo')
axarr[0].plot([-ksi2*wn2, -ksi2*wn2 ], [wn2*(1-ksi2**2)**0.5, -wn2*(1-ksi2**2)**0.5], 'ro')
axarr[0].set_title('Pole Positions')
axarr[0].axis([-3, 0.1, -3, 3])
axarr[0].grid()
axarr[0].set_ylabel('Imaginary')
axarr[0].set_xlabel('Real')
axarr[0].legend(['system-1', 'system-2'])
plt.show()
```
